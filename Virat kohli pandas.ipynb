```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year ?',y='Runs',hue='Format', palette='bright',marker='o')
plt.title("Total Run Per Year By Format")
plt.grid(True)
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(data=df, x='Format',y='Avg',hue='Format')
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f')
plt.title("Average Runs By format")

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='SR',y='Avg', hue='Format',size='Runs',sizes=(50,400))
plt.title("Strike Rate VS batting Avg")
plt.ylabel("Average")
plt.xlabel("Strike Rate")
```

```python
plt.figure(figsize=(8,6))
ax=sns.barplot(data=df,x='Format',y='100',estimator='sum',color='blue',label='100')
sns.barplot(data=df, x='Format',y='50',estimator='sum',color='skyblue',alpha=0.70,label='50')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
plt.legend()
plt.title("Total 100 and 50 by Format")
    
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=df, x='Year ?',y='SR',hue='Format',marker='o')
plt.title("Year Wise Strike Rate Trend")
plt.grid(True)
```

```python
df['Total_Boundaries']=df['4s']+df['6s']
plt.figure(figsize=(8,6))
ax=sns.barplot(data=df, x='Format',y='Total_Boundaries',estimator='sum',hue='Format')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Total Boundaries")
```

```python

plt.figure(figsize=(7,5))
sns.scatterplot(x=df["Runs"], y=df["SR"], hue=df["Year ?"], s=150)
plt.title("Runs vs Strike Rate")
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.show()

```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="Dot %", y="SR", hue="Year ?", s=150)
plt.title("Dot Ball % vs Strike Rate")
plt.show()
```
